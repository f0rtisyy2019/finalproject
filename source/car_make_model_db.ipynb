```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
model = df['Model'].unique().tolist()
print(len(model))
```

```python
make = df['Make'].unique().tolist()
make
```

```python
make_model = {}
for ma in make:
    make_model[ma] = []

print(make_model)
```

```python
for mo in model:
#     print(mo)
    try :
        make_key = df[df['Model'] == mo]['Make'].values[0]
#         print(make_key)
        if make_key in make_model:
            make_model[make_key].append(mo)
    except:
        print("Error")
#     print(make_model)
print(make_model)
```
